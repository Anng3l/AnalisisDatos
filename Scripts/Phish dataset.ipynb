```python
import pandas as pd
import pymongo
import re
```

```python
#Acceso a la BD de mongo y la colecci√≥n
my_cliente = pymongo.MongoClient("mongodb://localhost:27017/")
my_db = my_cliente["dataset"]
my_col = my_db["l"]
```

```python
df = pd.DataFrame(list(my_col.find()))
df
```

```python
#Crea una lista con las columnas a modificar
columnas = [
    'adj_gap', 'debut_dummy', 'notes_dummy', 'gap', 
    'total_times_played', 'set'
]
```

```python
#Conviere a num√©ricos los valores de las columnas establecidas en la lista. El error hace que si el valor no se puede
#transformar a num√©rico, se lo reemplace con NaN.
for col in columnas:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#Se reemplazan los NaN con 0 en las columas de la lista 'columnas'
df[columnas] = df[columnas].fillna(0)
```

```python
df
```

```python
#Convertir el Dataframe en una lista de diccionarios
data_dict = df.to_dict("records")
```

```python
#Eliminar todos los registros en la colecci√≥n
my_col.delete_many({})
```

```python
#Insertar los datos del dataframe actualizado
if data_dict:
    my_col.insert_many(data_dict)
```

```python
def clean_notes(note):
    if pd.isna(note):
        return ''
    # Eliminar \n, \t y espacios vac√≠os. re.sub(pattern, repl, string)
    cleaned_note = re.sub(r'\s+', ' ', note)  # Reemplaza m√∫ltiples espacios, \n y \t con un solo espacio
    cleaned_note = cleaned_note.strip()  # Elimina todos espacios al principio y al final
    return cleaned_note
```

```python
#Se aplica la funci√≥n por cada uno de los registros del campo 'notes'
df['notes'] = df['notes'].apply(clean_notes)
```

```python
df
```

```python
#Creaci√≥n de una lista de diccionario con los valores de df, eliminaci√≥n de los registros de la colecci√≥n y posterior inserci√≥n de los datos de
#los datos del diccionario hacia la colecci√≥n de la BD.
data_dict = df.to_dict("records")
my_col.delete_many({})
if data_dict:
    my_col.insert_many(data_dict)
```

```python
df
```

```python
#Reemplaza el string NA por NaN
df['notes'] = df['notes'].replace('NA', pd.NA)
```

```python
# Funci√≥n para reemplazar valores vac√≠os y NA con un texto gen√©rico
def replace_empty_and_na(note):
    if pd.isna(note) or note.strip() == '':
        return 'Informaci√≥n no disponible'  # Reemplazar con texto quemado si es NA o vac√≠o
    return note
```

```python
# Aplicar la funci√≥n para reemplazar valores vac√≠os y NA en la columna 'notes'
df['notes'] = df['notes'].apply(replace_empty_and_na)
```
